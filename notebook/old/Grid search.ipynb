```python
listeLRBF = [2,4,6,8]
listeNBFeatures = [50,100,150]
listeWeight = [0.01,0.05,0.1,0.15]
listeAlpha = [0,1,5,10]
listeRansac = [1,3,5,7]




lenRBF =len(listeLRBF)
lenNBFeatures = len(listeNBFeatures)
lenWeight = len(listeWeight)
lenAlpha = len(listeAlpha)
lenRansac = len(listeRansac)
```
