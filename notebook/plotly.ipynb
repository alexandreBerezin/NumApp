```python
import plotly.express as px
import numpy as np
```

```python
listeLRBF = [2,4,6,8]
listeNBFeatures = [50,100,150]
listeWeight = [0.01,0.05,0.1,0.15]
listeAlpha = [0,1,5,10]
listeRansac = [1,3,5,7]




lenRBF =len(listeLRBF)
lenNBFeatures = len(listeNBFeatures)
lenWeight = len(listeWeight)
lenAlpha = len(listeAlpha)
lenRansac = len(listeRansac)
```

```python
nbCol = 6
nbRow = lenRBF*lenNBFeatures*lenWeight*lenAlpha*lenRansac
```

```python
data = np.zeros((nbRow,nbCol))
```

```python
grid = np.load("gridSearch2.npy")
```

```python
data = np.zeros((nbRow,nbCol))
idx = 0 
for i in range(lenRBF):
    for j in range(lenNBFeatures):
        for k in range(lenWeight):
            for l in range(lenAlpha):
                for m in range(lenRansac):
                    
                    lRBF = listeLRBF[i]
                    nb = listeNBFeatures[j]
                    w = listeWeight[k]
                    alpha = listeAlpha[l]
                    reprojThreshold = listeRansac[m]
                    
                    ctr = grid[i,j,k,l,m]
                    
                    data[idx] = [lRBF,nb,w,alpha,reprojThreshold,ctr]
                    
                    idx = idx+1
                    
                    
```

```python
data
```

```python

dict = {
    "lRBF": data[:,0],
    "nbFeatures": data[:,1],
    "weightTVDeNois":data[:,2],
    "alpha":data[:,3],
    "threshold":data[:,4],
    "contraste":data[:,5],
    
    
}
fig = px.parallel_coordinates(dict,color="contraste",color_continuous_scale=px.colors.sequential.Jet)
fig.show()
```

```python
fig = px.scatter_matrix(dict,color = "contraste",color_continuous_scale=px.colors.sequential.Jet)
fig.show()
```
